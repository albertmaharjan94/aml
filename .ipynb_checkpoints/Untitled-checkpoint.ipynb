```python
songs_df = df2.copy()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.model_selection import train_test_split

songs_df = songs_df.dropna(subset=["track_popularity"])

features = ['acousticness', 'danceability', 'energy', 'valence']

songs_df = songs_df.dropna(subset=features)

song_sample = songs_df.sample(frac= 0.01)

X = song_sample[features]
y = song_sample['track_popularity']

X_normalized = (X - X.mean()) / X.std()

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.33, random_state=42)
```

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gpr = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=1)
gpr.fit(X_train, y_train)
```

```python
y_pred = gpr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R-squared:", r2)
```

```python
from sklearn.gaussian_process import GaussianProcessClassifier
y_classified = (y_train > y_train.median()).astype(int)
gpc = GaussianProcessClassifier(kernel=kernel)
gpc.fit(X_train, y_classified)
```

```python
y_predgpc = gpc.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_predgpc)
mse = mean_squared_error(y_test, y_predgpc)
r2 = r2_score(y_test, y_predgpc)

print("MAE:", mae)
print("MSE:", mse)
print("R-squared:", r2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

features = ['acousticness', 'danceability', 'energy', 'valence', 'tempo', 'track_popularity']
songs_df_selected = songs_df[features]

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
songs_df_imputed = imputer.fit_transform(songs_df_selected)
songs_df_imputed = np.nan_to_num(songs_df_imputed)
scaler = StandardScaler()
songs_df_scaled = scaler.fit_transform(songs_df_imputed)

pca = PCA(n_components=0.95) 
songs_df_reduced = pca.fit_transform(songs_df_scaled)

acousticness = ctrl.Antecedent(np.arange(0, 1, 0.01), 'acousticness')
danceability = ctrl.Antecedent(np.arange(0, 1, 0.01), 'danceability')

acousticness.automf(3) 
danceability.automf(3) 

recommendation = ctrl.Consequent(np.arange(0, 11, 1), 'recommendation')
recommendation.automf(3)  

#  fuzzy rules
rule1 = ctrl.Rule(acousticness['poor'] | danceability['good'], recommendation['good'])
rule2 = ctrl.Rule(acousticness['average'], recommendation['average'])
rule3 = ctrl.Rule(acousticness['good'] | danceability['poor'], recommendation['poor'])

# fuzzy control system
recommendation_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
recommendation_sim = ctrl.ControlSystemSimulation(recommendation_ctrl)

# setting input values
recommendation_sim.input['acousticness'] = 0.5 
recommendation_sim.input['danceability'] = 0.6  

# Compute 
recommendation_sim.compute()
print("Recommended Score: ", recommendation_sim.output['recommendation'])

# cosine similarity in batches 
batch_size = 1000 
similarity_matrix_batch = cosine_similarity(songs_df_reduced[:batch_size])

#  songs based on similarity
#  index of a song
song_index = 0 
similar_songs = similarity_matrix_batch[song_index]
recommended_song_indices = np.argsort(similar_songs)[::-1][1:]

print("Recommended Songs: ", recommended_song_indices)
```

```python
# fuzzy logic to the dataset
fuzzy_scores = []
for _, row in songs_df.iterrows():
    recommendation_sim.input['acousticness'] = row['acousticness']
    recommendation_sim.input['danceability'] = row['danceability']
    recommendation_sim.compute()
    fuzzy_scores.append(recommendation_sim.output['recommendation'])

# fuzzy scores to the data
songs_df['fuzzy_score'] = fuzzy_scores

# sortin
recommended_songs = songs_df.sort_values(by='fuzzy_score', ascending=False)
top_recommendations = recommended_songs.head(10)  # Top 10 recommendations
```

```python
plt.hist(songs_df['fuzzy_score'], bins=20, edgecolor='black')
plt.title('Distribution of Fuzzy Scores')
plt.xlabel('Fuzzy Score')
plt.ylabel('Frequency')
plt.show()
```

```python
for album_type in songs_df['album_type'].unique():
    subset = songs_df[songs_df['album_type'] == album_type]
    plt.hist(subset['fuzzy_score'], bins=20, alpha=0.5, label=str(album_type))
plt.title('Fuzzy Score Distribution by Album Type')
plt.xlabel('Fuzzy Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
